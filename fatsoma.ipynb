```python
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup

def extract_event_data(url):
    # Set up Firefox options
    firefox_options = Options()
    firefox_options.add_argument("--headless")
    firefox_options.add_argument("--no-sandbox")
    firefox_options.add_argument("--disable-dev-shm-usage")

    # Initialize the WebDriver
    driver = webdriver.Firefox(options=firefox_options)
    driver.get(url)
    time.sleep(5)  # Allow time for the page to load fully

    # Parse the fully loaded webpage with BeautifulSoup
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all event containers
    events = soup.find_all('div', class_='_variant-default_1jkwo3')

    # Extract data for each event
    event_data = []
    seen_events = set()  # To track unique events

    for event in events:
        try:
            name_element = event.find('a', class_='_name_1jkwo3')
            name = name_element.get_text(strip=True)
            event_url = name_element['href'] if name_element else ""
            date = event.find('div', class_='_date_1jkwo3').get_text(strip=True)
            location = event.find('div', class_='_location_1jkwo3').get_text(strip=True)
            price = event.find('div', class_='_price_1jkwo3').get_text(strip=True)

            # Safely extract image URL
            image_element = event.find('img', class_='_img-bg_1jkwo3')
            image_url = image_element.get('src', '')  # Safely get the 'src' attribute

            # Create a unique identifier for the event
            event_id = f"{name}-{date}"

            # Add to the list only if it's not a duplicate
            if event_id not in seen_events:
                seen_events.add(event_id)  # Mark this event as seen
                event_data.append({
                    'Event Name': name,
                    'Date and Time': date,
                    'Location': location,
                    'Price': price,
                    'Image URL': image_url,
                    'URL': f"https://www.fatsoma.com{event_url}",
                    'Category': 'Sports and Recreation'
                })
        except AttributeError:
            # Skip events with missing data
            continue

    # Close the browser
    driver.quit()

    # Save extracted data to a JSON file
    with open('sr_data.json', 'w', encoding='utf-8') as json_file:
        json.dump(event_data, json_file, ensure_ascii=False, indent=4)

    print("Event data has been saved to events_data.json")
    return event_data

# URL to process
URL = 'https://www.fatsoma.com/discover?category-id=bcd84c45-a1cb-4658-b3f4-601fda4a1c6a'

# Extract data and save to JSON file
event_data = extract_event_data(URL)
print(event_data)
```
