```python
import time
import json
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException

def extract_property_data(urls):
    # Set up Firefox options
    firefox_options = Options()
    firefox_options.add_argument("--headless")  # Run in headless mode
    firefox_options.add_argument("--no-sandbox")
    firefox_options.add_argument("--disable-dev-shm-usage")

    # Initialize the WebDriver
    driver = webdriver.Firefox(options=firefox_options)

    properties = []

    for url in urls:
        # Remove duplicate segments in URL if necessary
        url = url.replace("https://amberstudent.comhttps://amberstudent.com", "https://amberstudent.com")

        driver.get(url)
        time.sleep(5)  # Allow time for page to load fully

        try:
            # Wait for and get property details
            room_name_element = WebDriverWait(driver, 20).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, "h1[data-testid='propertypage-property-name']"))
            )
            room_name = room_name_element.text

            # Extract other property details
            location = driver.find_element(By.CSS_SELECTOR, "div.PropertyDetailsSectionStyles-module__propertyAddress").text
            cost = driver.find_element(By.CSS_SELECTOR, "div.PropertyDetailsSectionStyles-module__detailHeaderRhs").text
            distance = driver.find_element(By.CSS_SELECTOR, "div.PropertyDetailsSectionStyles-module__distanceFromCenter").text
            walking_distance = driver.find_element(By.CSS_SELECTOR, "span[data-testid='searchpage-inventoryCard-walkingDistance']").text
            train_distance = driver.find_element(By.CSS_SELECTOR, "span[data-testid='searchpage-inventoryCard-trainDistance']").text
            driving_distance = driver.find_element(By.CSS_SELECTOR, "span[data-testid='searchpage-inventoryCard-drivingDistance']").text

            tags = [tag.text for tag in driver.find_elements(By.CSS_SELECTOR, "div.ScrollStyles-module__cityName")]
            offers = [offer.text for offer in driver.find_elements(By.CSS_SELECTOR, "div[data-testid='property-policy-name']")]
            description = driver.find_element(By.CSS_SELECTOR, "div[data-testid='property-detail-description']").text

            # Room details
            room_details = []
            room_cards = driver.find_elements(By.CSS_SELECTOR, "div[data-testid='room-types-card']")
            for room_card in room_cards:
                try:
                    room_type_name = room_card.find_element(By.CSS_SELECTOR, "h4[data-testid='propety-room-type-name']").text
                    starting_price = room_card.find_element(By.CSS_SELECTOR, "span[data-testid='property-room-type-price-range']").text.replace("Starting From: ", "").strip()
                    availability_date = room_card.find_element(By.CSS_SELECTOR, "span[data-testid='property-room-type-availability-from']").text
                    features = [feature.text for feature in room_card.find_elements(By.CSS_SELECTOR, "div[data-testid='property-room-type-tenancy-amenities']")]


                    room_details.append({
                        "room_type_name": room_type_name,
                        "starting_price": starting_price,
                        "availability_date": availability_date,
                        "features": features

                    })
                except NoSuchElementException:
                    print("Some room details not found, skipping this room.")

            # Append data with city information
            p_data = {
                "room_name": room_name,
                "city": "Nottingham",
                "location": location,
                "cost": cost,
                "distance": distance,
                "walking_distance": walking_distance,
                "train_distance": train_distance,
                "driving_distance": driving_distance,
                "tags": tags,
                "offers": offers,
                "description": description,
                "room_details": room_details
            }
            properties.append(p_data)

        except (NoSuchElementException, TimeoutException) as e:
            print(f"Error extracting data from {url}: {str(e)}")

    driver.quit()

    # Save data to JSON file
    with open("nottingham_details.json", "w", encoding="utf-8") as f:
        json.dump(properties, f, indent=2, ensure_ascii=False)

# List of URLs to process
urls = l
# Extract data and save to JSON file
extract_property_data(urls)
print("Data saved to properties_data.json")
```
