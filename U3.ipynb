```python
import requests
from bs4 import BeautifulSoup
import json

# Base URL for the search page with pagination
base_url = "https://www.ucas.com/explore/search/providers?query=&page="

# Function to extract university data from a page
def extract_university_data(page_url):
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # List to store university data
    universities = []

    # Find all university entries
    university_blocks = soup.find_all('div', class_='content__details')

    for block in university_blocks:
        # Extract university name
        name = block.find('p', class_='header__text').text.strip() if block.find('p', class_='header__text') else 'N/A'

        # Extract location
        location = block.find('p', class_='location-display__location').text.strip() if block.find('p', class_='location-display__location') else 'N/A'

        # Extract background image URL
        image_block = block.find_previous('div', class_='content__image')
        background_image = image_block['data-lazy-load-content'] if image_block and 'data-lazy-load-content' in image_block.attrs else 'N/A'

        # Extract logo image URL
        logo_block = image_block.find('img') if image_block else None
        logo_image = logo_block['data-url'] if logo_block and 'data-url' in logo_block.attrs else 'N/A'

        # Store data in dictionary
        university_data = {
            'name': name,
            'location': location,
            'background_image': background_image,
            'logo_image': logo_image
        }

        # Append to the list
        universities.append(university_data)

    return universities

# Function to scrape university data from multiple pages
def scrape_universities(max_pages):
    all_universities = []

    # Loop through each page number, constructing the URL dynamically
    for page in range(1, max_pages + 1):
        current_url = base_url + str(page)
        print(f"Scraping page {page}: {current_url}")

        # Extract data from the current page
        universities = extract_university_data(current_url)
        all_universities.extend(universities)

    return all_universities

# Specify the maximum number of pages to scrape (set to 22 or higher)
max_pages = 22  # You can adjust this based on the number of available pages

# Start scraping data from the specified number of pages
all_university_data = scrape_universities(max_pages)

# Convert the data to JSON format
universities_json = json.dumps(all_university_data, indent=4)

# Print the JSON output
print(universities_json)

# Optionally, save to a JSON file
with open('universities_paginated.json', 'w') as file:
    file.write(universities_json)

print(f"Scraping completed successfully. Data saved to 'universities_paginated.json'.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# String containing all university URLs, separated by newlines
university_urls = """https://www.ucas.com/explore/unis/3d3a3bb5/acm-the-academy-of-contemporary-music
https://www.ucas.com/explore/unis/4e978513/arts-university-bournemouth
https://www.ucas.com/explore/unis/3824407c/associated-studios-performing-arts-academy
https://www.ucas.com/explore/unis/b7599006/university-college-birmingham
https://www.ucas.com/explore/unis/898276b4/bloomsbury-institute-london
https://www.ucas.com/explore/unis/627021f2/university-of-bradford
https://www.ucas.com/explore/unis/27ea792a/university-of-brighton
https://www.ucas.com/explore/unis/b79a4cbd/bristol-university-of-the-west-of-england
https://www.ucas.com/explore/unis/42121682/brunel-university-london
https://www.ucas.com/explore/unis/dfed9a6a/coventry-university
https://www.ucas.com/explore/unis/1de3729b/university-for-the-creative-arts
https://www.ucas.com/explore/unis/008ac8a8/university-of-derby
https://www.ucas.com/explore/unis/44f558c0/edge-hill-university
https://www.ucas.com/explore/unis/e4c66810/university-of-glasgow
https://www.ucas.com/explore/unis/e2cc6194/university-of-hertfordshire
https://www.ucas.com/explore/unis/c1c7b933/university-of-the-highlands-and-islands-uhi
https://www.ucas.com/explore/unis/31d5869a/hult-international-business-school
https://www.ucas.com/explore/unis/52087408/imperial-college-london
https://www.ucas.com/explore/unis/6cadf6e5/the-university-of-law
https://www.ucas.com/explore/unis/edaad69a/university-of-leicester
https://www.ucas.com/explore/unis/96faf64f/libf
https://www.ucas.com/explore/unis/970c6a07/university-of-lincoln
https://www.ucas.com/explore/unis/3e00a545/london-interdisciplinary-school-lis
https://www.ucas.com/explore/unis/bc261a05/mctimoney-college-of-chiropractic
https://www.ucas.com/explore/unis/5692045c/middlesex-university
https://www.ucas.com/explore/unis/9528a5b4/newcastle-university
https://www.ucas.com/explore/unis/1f6ba907/university-of-northampton
https://www.ucas.com/explore/unis/9698b0cc/nottingham-trent-university
https://www.ucas.com/explore/unis/79f42328/university-of-nottingham
https://www.ucas.com/explore/unis/a3bf2ac2/plymouth-marjon-university
https://www.ucas.com/explore/unis/250ab332/university-of-plymouth
https://www.ucas.com/explore/unis/ee9a8e04/pointblank-music-school
https://www.ucas.com/explore/unis/712655f4/richmond-american-university-london
https://www.ucas.com/explore/unis/895f238c/university-of-roehampton
https://www.ucas.com/explore/unis/f2b98edd/royal-veterinary-college-university-of-london
https://www.ucas.com/explore/unis/a33ecb69/university-of-south-wales
https://www.ucas.com/explore/unis/0f5d882f/university-of-southampton
https://www.ucas.com/explore/unis/f1f9a1ea/sp-jain-london-school-of-management
https://www.ucas.com/explore/unis/cf271b96/university-of-suffolk
https://www.ucas.com/explore/unis/6e856d76/university-of-sussex
https://www.ucas.com/explore/unis/4b70e003/swansea-university
https://www.ucas.com/explore/unis/62431d8f/university-of-worcester
https://www.ucas.com/explore/unis/3a023828/university-of-aberdeen
https://www.ucas.com/explore/unis/1558119a/abertay-university
https://www.ucas.com/explore/unis/168d2cd8/aberystwyth-university
https://www.ucas.com/explore/unis/a04f3f5e/academy-of-live-technology
https://www.ucas.com/explore/unis/1630c359/activate-learning
https://www.ucas.com/explore/unis/57f3e439/amity-university-in-london
https://www.ucas.com/explore/unis/d890f3a3/anglia-ruskin-university
https://www.ucas.com/explore/unis/15f49bd0/architectural-association-school-of-architecture
https://www.ucas.com/explore/unis/8554563c/arden-university
https://www.ucas.com/explore/unis/d4a8f435/arts-educational-schools
https://www.ucas.com/explore/unis/fd337b6e/arts-university-plymouth
https://www.ucas.com/explore/unis/95371765/aru-london
https://www.ucas.com/explore/unis/7f5fefd6/aru-writtle-formerly-writtle-university-college
https://www.ucas.com/explore/unis/7d51ec43/university-centre-askham-bryan
https://www.ucas.com/explore/unis/89c29ffe/aston-university-birmingham
https://www.ucas.com/explore/unis/7522cc92/bangor-university
https://www.ucas.com/explore/unis/a24b2b81/barnet-and-southgate-college
https://www.ucas.com/explore/unis/5f7d6526/barnfield-college-luton
https://www.ucas.com/explore/unis/4f52efe2/barnsley-college-university-centre
https://www.ucas.com/explore/unis/4c5e93f0/bath-college
https://www.ucas.com/explore/unis/29a67da9/bath-spa-university
https://www.ucas.com/explore/unis/295cc6a3/university-of-bath
https://www.ucas.com/explore/unis/a6fa6f7f/bedford-college-group
https://www.ucas.com/explore/unis/3261960f/university-of-bedfordshire
https://www.ucas.com/explore/unis/a537a0df/belfast-school-of-theology
https://www.ucas.com/explore/unis/5871db48/bimm-university
https://www.ucas.com/explore/unis/a5e9ffc8/birkbeck-university-of-london
https://www.ucas.com/explore/unis/24d90a06/birmingham-city-university
https://www.ucas.com/explore/unis/fa8f4026/bmet-birmingham-metropolitan-college
https://www.ucas.com/explore/unis/11ac4ef6/birmingham-newman-university
https://www.ucas.com/explore/unis/1d61dc76/university-of-birmingham
https://www.ucas.com/explore/unis/2519869b/university-centre-bishop-burton
https://www.ucas.com/explore/unis/a734fa1a/bishop-grosseteste-university
https://www.ucas.com/explore/unis/c2fe0703/blackburn-college
https://www.ucas.com/explore/unis/15d7c39f/blackpool-and-the-fylde-college
https://www.ucas.com/explore/unis/88e1a042/university-of-bolton
https://www.ucas.com/explore/unis/96f866ad/boomsatsuma
https://www.ucas.com/explore/unis/61cf12e4/boston-college
https://www.ucas.com/explore/unis/d4b5280f/bournemouth-and-poole-college
https://www.ucas.com/explore/unis/22e93c90/bournemouth-university
https://www.ucas.com/explore/unis/95d4242b/bpp-university
https://www.ucas.com/explore/unis/90e9d147/bradford-college
https://www.ucas.com/explore/unis/d849a737/bridgend-college
https://www.ucas.com/explore/unis/93e3e3b3/university-centre-somerset---bridgwater-taunton-college
https://www.ucas.com/explore/unis/0bdd99ec/brighton-and-sussex-medical-school
https://www.ucas.com/explore/unis/692982bf/city-of-bristol-college
https://www.ucas.com/explore/unis/da03c49a/trinity-college-bristol
https://www.ucas.com/explore/unis/69690f14/university-of-bristol
https://www.ucas.com/explore/unis/4cefe4e9/british-academy-of-jewellery
https://www.ucas.com/explore/unis/98e33cb7/bcom-eso---part-of-bcno-group-london-kent
https://www.ucas.com/explore/unis/1386af38/university-of-buckingham
https://www.ucas.com/explore/unis/592fda96/buckinghamshire-new-university
https://www.ucas.com/explore/unis/b0175104/burnley-college
https://www.ucas.com/explore/unis/74de5b4c/bury-college
https://www.ucas.com/explore/unis/27dbeccf/cambridge-marketing-college
https://www.ucas.com/explore/unis/3c88a10a/cambridge-regional-college
https://www.ucas.com/explore/unis/83b36cda/cambridge-school-of-visual-and-performing-arts
https://www.ucas.com/explore/unis/badc3603/university-of-cambridge
https://www.ucas.com/explore/unis/24bec0e2/canterbury-christ-church-university
https://www.ucas.com/explore/unis/6a3d68d5/cardiff-and-vale-college
https://www.ucas.com/explore/unis/777de2ea/cardiff-metropolitan-university
https://www.ucas.com/explore/unis/e1c1ad69/cardiff-university
https://www.ucas.com/explore/unis/016d384d/central-film-school-london
https://www.ucas.com/explore/unis/14062c40/university-of-central-lancashire-london
https://www.ucas.com/explore/unis/6533bebc/university-of-central-lancashire
https://www.ucas.com/explore/unis/81f409fb/centre-for-alternative-technology
https://www.ucas.com/explore/unis/09d21dac/university-of-chester
https://www.ucas.com/explore/unis/02675d55/chesterfield-college
https://www.ucas.com/explore/unis/52535f33/chichester-college-group-inc-brinsbury-chichester-crawley-northbrook-worthing-colleges
https://www.ucas.com/explore/unis/b1f140f3/university-of-chichester
https://www.ucas.com/explore/unis/bb82986c/city-guilds-of-london-art-school
https://www.ucas.com/explore/unis/e0a6cb6c/city-college-norwich-incorporating-easton-college
https://www.ucas.com/explore/unis/0e86a0cf/city-university-of-london
https://www.ucas.com/explore/unis/8ff764e4/cliff-college
https://www.ucas.com/explore/unis/c58b1306/university-centre-colchester-at-colchester-institute
https://www.ucas.com/explore/unis/12a66ed8/grwp-llandrillo-menai
https://www.ucas.com/explore/unis/1729a34a/collective-acting-studio
https://www.ucas.com/explore/unis/eaafa9fa/college-of-agriculture-food-and-rural-enterprise
https://www.ucas.com/explore/unis/878ef093/college-of-esports
https://www.ucas.com/explore/unis/a435d80e/cond-nast-college-of-fashion-design
https://www.ucas.com/explore/unis/6fc7541d/cornwall-college
https://www.ucas.com/explore/unis/6ec9c892/courtauld-institute-of-art-university-of-london
https://www.ucas.com/explore/unis/e86a79e3/coventry-college
https://www.ucas.com/explore/unis/2a9fba46/cranfield-university
https://www.ucas.com/explore/unis/aecda516/craven-college
https://www.ucas.com/explore/unis/ca2a0ff9/croydon-university-centre
https://www.ucas.com/explore/unis/ee20f1e0/university-of-cumbria
https://www.ucas.com/explore/unis/e859e326/david-game-college---higher-education-centre
https://www.ucas.com/explore/unis/11b2cb24/dbs-institute
https://www.ucas.com/explore/unis/6d06ed34/de-montfort-university
https://www.ucas.com/explore/unis/1fd8988e/derby-college
https://www.ucas.com/explore/unis/a3e4c2d4/dn-colleges-group
https://www.ucas.com/explore/unis/0299fcf2/dudley-college
https://www.ucas.com/explore/unis/bfcb0d31/university-of-dundee
https://www.ucas.com/explore/unis/6e455568/durham-university
https://www.ucas.com/explore/unis/2daba510/new-college-durham
https://www.ucas.com/explore/unis/70100535/dyslexia-action
https://www.ucas.com/explore/unis/7dce1b37/east-15-acting-school
https://www.ucas.com/explore/unis/59590204/university-of-east-anglia-uea
https://www.ucas.com/explore/unis/ae6f4d85/ekc-group
https://www.ucas.com/explore/unis/6d89f913/university-of-east-london
https://www.ucas.com/explore/unis/7837e9ef/east-riding-college
https://www.ucas.com/explore/unis/9597b298/east-sussex-college
https://www.ucas.com/explore/unis/dc3ccbfc/edinburgh-napier-university
https://www.ucas.com/explore/unis/114bb2e7/the-university-of-edinburgh
https://www.ucas.com/explore/unis/c615f773/escape-studios
https://www.ucas.com/explore/unis/24524819/escp-business-school
https://www.ucas.com/explore/unis/91f1f141/university-of-essex
https://www.ucas.com/explore/unis/ed93b6a5/european-school-of-economics
https://www.ucas.com/explore/unis/1f21f18b/exeter-college
https://www.ucas.com/explore/unis/91fd2fa0/university-of-exeter
https://www.ucas.com/explore/unis/13fe6f4f/falmouth-university
https://www.ucas.com/explore/unis/56ce8405/university-centre-farnborough
https://www.ucas.com/explore/unis/9a25a2cc/fashion-retail-academy
https://www.ucas.com/explore/unis/e1e76de4/furness-college
https://www.ucas.com/explore/unis/1018280f/futureworks
https://www.ucas.com/explore/unis/2304d818/gateshead-college
https://www.ucas.com/explore/unis/5cffcbbb/glasgow-caledonian-university
https://www.ucas.com/explore/unis/dd2bb28c/city-of-glasgow-college
https://www.ucas.com/explore/unis/a7656ba9/glasgow-school-of-art
https://www.ucas.com/explore/unis/098e22b9/gloucestershire-college
https://www.ucas.com/explore/unis/aa1a1601/university-of-gloucestershire
https://www.ucas.com/explore/unis/a8aae742/goldsmiths-university-of-london
https://www.ucas.com/explore/unis/d36d615d/gower-college-swansea
https://www.ucas.com/explore/unis/2c82b225/university-of-greenwich
https://www.ucas.com/explore/unis/5e389d01/guildhall-school-of-music-drama
https://www.ucas.com/explore/unis/87fbf767/halesowen-college
https://www.ucas.com/explore/unis/c9a6da19/harlow-college
https://www.ucas.com/explore/unis/a249ae49/harper-adams-university
https://www.ucas.com/explore/unis/788cb34b/harper-and-keele-veterinary-school
https://www.ucas.com/explore/unis/abf8e17f/hartpury-university
https://www.ucas.com/explore/unis/3ed8539f/havant-and-south-downs-college
https://www.ucas.com/explore/unis/9bca8f42/new-city-college-incorporating-havering-college
https://www.ucas.com/explore/unis/e9d03a65/haybridge-alliance-school-centred-initial-teacher-training
https://www.ucas.com/explore/unis/25f59ca8/harrow-richmond-uxbridge-college
https://www.ucas.com/explore/unis/d920c4e2/health-sciences-university-formerly-aecc-university-college
https://www.ucas.com/explore/unis/48a4bc2c/heart-of-worcestershire-college
https://www.ucas.com/explore/unis/ebb4f05e/hereford-college-of-arts
https://www.ucas.com/explore/unis/09715d18/herefordshire-ludlow-north-shropshire-college
https://www.ucas.com/explore/unis/5c658fd1/heriot-watt-university
https://www.ucas.com/explore/unis/73efa26e/hertford-regional-college
https://www.ucas.com/explore/unis/9ccfa755/holy-cross-sixth-form-college-and-university-centre
https://www.ucas.com/explore/unis/32bce31b/hopwood-hall-college
https://www.ucas.com/explore/unis/c666f13c/university-of-huddersfield
https://www.ucas.com/explore/unis/1231e6ce/hugh-baird-college
https://www.ucas.com/explore/unis/cbb9695b/hull-college
https://www.ucas.com/explore/unis/6a15243b/hull-york-medical-school
https://www.ucas.com/explore/unis/ac1fb4b5/university-of-hull
https://www.ucas.com/explore/unis/4215e8c6/icon-college-of-technology-and-management
https://www.ucas.com/explore/unis/9fd86a3e/institute-of-advanced-legal-studies-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/44eed984/institute-of-cancer-research-university-of-london
https://www.ucas.com/explore/unis/d04f617a/institute-of-commonwealth-studies-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/b829718a/institute-of-contemporary-music-performance
https://www.ucas.com/explore/unis/5737f385/institute-of-development-studies
https://www.ucas.com/explore/unis/909e0405/institute-of-english-studies-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/3e0010c4/institute-of-historical-research-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/08f1a8cc/institute-of-languages-cultures-and-societies-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/54791c19/international-business-college-manchester
https://www.ucas.com/explore/unis/16963c21/international-college-of-musical-theatre-icmt
https://www.ucas.com/explore/unis/a96bc33f/international-luxury-academy
https://www.ucas.com/explore/unis/c1eb45fd/irish-baptist-college
https://www.ucas.com/explore/unis/dd033310/islamic-college-for-advanced-studies
https://www.ucas.com/explore/unis/23aec7b0/istituto-marangoni-london
https://www.ucas.com/explore/unis/bb956d7d/jca-london-fashion-academy
https://www.ucas.com/explore/unis/5e07e833/keele-university
https://www.ucas.com/explore/unis/99fa626a/kent-and-medway-medical-school
https://www.ucas.com/explore/unis/aa22c216/university-of-kent
https://www.ucas.com/explore/unis/3cb12f40/king-stage
https://www.ucas.com/explore/unis/5fd4b1cc/king-s-college-london-university-of-london
https://www.ucas.com/explore/unis/58bbc794/kingston-maurward-college
https://www.ucas.com/explore/unis/4d6287f3/kingston-university
https://www.ucas.com/explore/unis/eb187029/klc-school-of-design
https://www.ucas.com/explore/unis/73bc7b26/lamda-london-academy-of-music-dramatic-art
https://www.ucas.com/explore/unis/058545dd/lancaster-university
https://www.ucas.com/explore/unis/b79f887e/lccm
https://www.ucas.com/explore/unis/272be5f0/leeds-arts-university
https://www.ucas.com/explore/unis/059b84a8/leeds-beckett-university
https://www.ucas.com/explore/unis/fe767449/university-centre-leeds-leeds-city-college
https://www.ucas.com/explore/unis/74edaea7/leeds-conservatoire
https://www.ucas.com/explore/unis/10bb36d7/leeds-trinity-university
https://www.ucas.com/explore/unis/b03b9d06/university-of-leeds
https://www.ucas.com/explore/unis/082a9246/leicester-college
https://www.ucas.com/explore/unis/d680310d/lincoln-college-university-centre
https://www.ucas.com/explore/unis/bf0964d9/city-of-liverpool-college-university-centre
https://www.ucas.com/explore/unis/78c33534/liverpool-hope-university
https://www.ucas.com/explore/unis/0097ca6a/liverpool-institute-for-performing-arts
https://www.ucas.com/explore/unis/b4d790d4/liverpool-john-moores-university
https://www.ucas.com/explore/unis/ff27f6b2/liverpool-school-of-tropical-medicine
https://www.ucas.com/explore/unis/9a2e99c3/university-of-liverpool
https://www.ucas.com/explore/unis/812e7c31/lma
https://www.ucas.com/explore/unis/32d682dc/london-business-school-university-of-london
https://www.ucas.com/explore/unis/f0896af7/london-college
https://www.ucas.com/explore/unis/0115c340/london-college-of-business-studies
https://www.ucas.com/explore/unis/d9ea977d/london-college-of-contemporary-arts
https://www.ucas.com/explore/unis/436d2160/london-film-academy
https://www.ucas.com/explore/unis/04dc593c/london-film-school
https://www.ucas.com/explore/unis/36dc6feb/london-metropolitan-university
https://www.ucas.com/explore/unis/25fe3db7/london-school-of-economics-and-political-science-university-of-london
https://www.ucas.com/explore/unis/b137fc85/london-school-of-hygiene-tropical-medicine-university-of-london
https://www.ucas.com/explore/unis/6ac2c56a/london-school-of-management-education
https://www.ucas.com/explore/unis/73c1c444/london-school-of-theology
https://www.ucas.com/explore/unis/b8a3b6cb/london-south-bank-university
https://www.ucas.com/explore/unis/aba77858/london-south-east-colleges-bromley-college
https://www.ucas.com/explore/unis/f838c19c/loughborough-college
https://www.ucas.com/explore/unis/b935bbb7/loughborough-university
https://www.ucas.com/explore/unis/bb9bf30f/manchester-metropolitan-university
https://www.ucas.com/explore/unis/01b03c44/university-of-manchester
https://www.ucas.com/explore/unis/e53c5903/medway-school-of-pharmacy
https://www.ucas.com/explore/unis/fa347416/metfilm
https://www.ucas.com/explore/unis/44c02f19/middlesbrough-college
https://www.ucas.com/explore/unis/7cb2e0fa/midkent-college-inc-medway-school-of-arts
https://www.ucas.com/explore/unis/a5ec7e73/morley-college-london
https://www.ucas.com/explore/unis/15cba371/moulton-college
https://www.ucas.com/explore/unis/3a59e89f/mountview-academy-of-theatre-arts
https://www.ucas.com/explore/unis/7d269cca/university-centre-myerscough
https://www.ucas.com/explore/unis/3886d908/national-college-of-hypnosis-psychotherapy
https://www.ucas.com/explore/unis/e0e252b0/national-film-and-television-school
https://www.ucas.com/explore/unis/a30c9a10/nazarene-theological-college
https://www.ucas.com/explore/unis/3a1c51f4/nelson-and-colne-college-group
https://www.ucas.com/explore/unis/cb222c5f/nescot
https://www.ucas.com/explore/unis/a524ae42/new-model-institute-for-technology-and-engineering-nmite
https://www.ucas.com/explore/unis/bbb76d07/new-school-of-psychotherapy-and-counselling
https://www.ucas.com/explore/unis/41110e8a/newbold-college
https://www.ucas.com/explore/unis/6783d4af/newcastle-college-university-centre
https://www.ucas.com/explore/unis/3cc3eeb1/newham-college-of-further-education
https://www.ucas.com/explore/unis/881106e7/norland
https://www.ucas.com/explore/unis/08b3f282/north-hertfordshire-college
https://www.ucas.com/explore/unis/aad4cda5/north-kent-college
https://www.ucas.com/explore/unis/ca9f4d85/north-warwickshire-and-south-leicestershire-college
https://www.ucas.com/explore/unis/77bfcfb8/northeastern-university-london
https://www.ucas.com/explore/unis/ad1ad199/northern-college-of-acupuncture
https://www.ucas.com/explore/unis/590e6636/the-northern-school-of-art
https://www.ucas.com/explore/unis/bc655e18/northern-school-of-contemporary-dance
https://www.ucas.com/explore/unis/408b9be4/northumbria-university-newcastle
https://www.ucas.com/explore/unis/bd68226f/norwich-university-of-the-arts
https://www.ucas.com/explore/unis/319ba5b1/nottingham-college
https://www.ucas.com/explore/unis/d1045d7f/nptc-group
https://www.ucas.com/explore/unis/4f7a0e47/oak-hill-theological-college
https://www.ucas.com/explore/unis/8c7662bf/oaklands-college
https://www.ucas.com/explore/unis/d2e1adfe/university-campus-oldham
https://www.ucas.com/explore/unis/7f559a95/open-university
https://www.ucas.com/explore/unis/d66b3801/orbital-south-colleges-university-centre
https://www.ucas.com/explore/unis/d422b186/oxford-brookes-university
https://www.ucas.com/explore/unis/351374e0/university-of-oxford
https://www.ucas.com/explore/unis/5986cf35/peter-symonds-college
https://www.ucas.com/explore/unis/7ca2e52f/university-centre-peterborough
https://www.ucas.com/explore/unis/fba7f8e7/petroc
https://www.ucas.com/explore/unis/be7280d5/plumpton-college
https://www.ucas.com/explore/unis/babaeca8/city-college-plymouth
https://www.ucas.com/explore/unis/039b8331/university-of-portsmouth
https://www.ucas.com/explore/unis/403c3afa/queen-margaret-university-edinburgh
https://www.ucas.com/explore/unis/856d62c2/queen-mary-university-of-london
https://www.ucas.com/explore/unis/3e45ece0/queen-s-university-belfast
https://www.ucas.com/explore/unis/6f003a01/ravensbourne-university-london
https://www.ucas.com/explore/unis/c283b1b0/university-of-reading
https://www.ucas.com/explore/unis/b3301895/real-training
https://www.ucas.com/explore/unis/ae8f02f9/university-centre-reaseheath
https://www.ucas.com/explore/unis/4687847f/refugee-law-initiative-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/d22aa302/regent-s-university-london
https://www.ucas.com/explore/unis/60144100/regents-theological-college
https://www.ucas.com/explore/unis/5a3cac22/robert-gordon-university
https://www.ucas.com/explore/unis/5bc57c84/rose-bruford-college
https://www.ucas.com/explore/unis/94090b0b/royal-academy-of-dance
https://www.ucas.com/explore/unis/ac1ac95e/royal-academy-of-dramatic-art
https://www.ucas.com/explore/unis/6650b619/royal-academy-of-music-university-of-london
https://www.ucas.com/explore/unis/0014c21f/royal-agricultural-university
https://www.ucas.com/explore/unis/e4015e94/royal-birmingham-conservatoire
https://www.ucas.com/explore/unis/098ac3d1/royal-central-school-of-speech-and-drama-university-of-london
https://www.ucas.com/explore/unis/b6e2e7ed/royal-college-of-art
https://www.ucas.com/explore/unis/5f0445dd/royal-college-of-music
https://www.ucas.com/explore/unis/22072435/royal-conservatoire-of-scotland
https://www.ucas.com/explore/unis/fb1d3c04/royal-drawing-school
https://www.ucas.com/explore/unis/5b13b3ae/royal-holloway-university-of-london
https://www.ucas.com/explore/unis/dd590a4b/royal-northern-college-of-music
https://www.ucas.com/explore/unis/76a6ba43/royal-welsh-college-of-music-and-drama
https://www.ucas.com/explore/unis/db037e28/sae-institute
https://www.ucas.com/explore/unis/81fe3292/university-of-st-andrews
https://www.ucas.com/explore/unis/9b3207d7/st-george-s-university-of-london
https://www.ucas.com/explore/unis/e47a9792/st-mary-s-university-college-belfast
https://www.ucas.com/explore/unis/8432fe0e/university-centre-at-salford-city-college
https://www.ucas.com/explore/unis/07fe4a70/university-of-salford
https://www.ucas.com/explore/unis/80031aa0/sgs-south-gloucestershire-and-stroud-college
https://www.ucas.com/explore/unis/86bcfbc8/sheffield-college
https://www.ucas.com/explore/unis/af611462/sheffield-hallam-university
https://www.ucas.com/explore/unis/dbd76d8b/university-of-sheffield
https://www.ucas.com/explore/unis/a840508b/shrewsbury-colleges-group
https://www.ucas.com/explore/unis/2a479d97/coleg-sir-gar
https://www.ucas.com/explore/unis/01d1372f/smb-college-group
https://www.ucas.com/explore/unis/63840f6d/soas-university-of-london
https://www.ucas.com/explore/unis/26360c43/solihull-college-university-centre
https://www.ucas.com/explore/unis/b532a77f/sotheby-s-institute-of-art-london
https://www.ucas.com/explore/unis/0c59f8f5/south-and-city-college-birmingham
https://www.ucas.com/explore/unis/b44ad48d/south-thames-college-group-inc-carshalton-south-thames-and-kingston-college
https://www.ucas.com/explore/unis/f92b2692/solent-university-southampton
https://www.ucas.com/explore/unis/e71cb3c4/university-centre-sparsholt
https://www.ucas.com/explore/unis/dc96ef3c/spurgeon-s-college
https://www.ucas.com/explore/unis/ba944ca6/sruc-scotland-s-rural-college
https://www.ucas.com/explore/unis/f827e91c/university-centre-st-helens
https://www.ucas.com/explore/unis/1264b52a/st-mary-s-university-twickenham
https://www.ucas.com/explore/unis/20b2137b/university-of-stirling
https://www.ucas.com/explore/unis/337ec3b8/stranmillis-university-college
https://www.ucas.com/explore/unis/5cc768aa/university-of-strathclyde
https://www.ucas.com/explore/unis/31260746/city-of-sunderland-college
https://www.ucas.com/explore/unis/5f968715/university-of-sunderland
https://www.ucas.com/explore/unis/5ea13bf3/university-of-surrey
https://www.ucas.com/explore/unis/b35862e3/the-college-swansea-university
https://www.ucas.com/explore/unis/16a7e353/new-college-swindon-university-centre
https://www.ucas.com/explore/unis/4156d693/tameside-college
https://www.ucas.com/explore/unis/bffe1f14/the-engineering-design-institute-london
https://www.ucas.com/explore/unis/b46d1424/teesside-university
https://www.ucas.com/explore/unis/6c26b7ba/thinkspace-education
https://www.ucas.com/explore/unis/ad154893/trafford-stockport-college-group
https://www.ucas.com/explore/unis/09c81910/trinity-laban
https://www.ucas.com/explore/unis/cfe1b520/university-centre-truro-and-penwith
https://www.ucas.com/explore/unis/117bff52/two-mile-ash-ittp-partnered-with-chiltern-training-group
https://www.ucas.com/explore/unis/8df995e9/tyne-coast-college
https://www.ucas.com/explore/unis/babb7b8c/ucen-manchester
https://www.ucas.com/explore/unis/9bebda25/ucfb
https://www.ucas.com/explore/unis/bd0c70d8/ucl-university-college-london
https://www.ucas.com/explore/unis/b0c96a45/ulster-university
https://www.ucas.com/explore/unis/2f918b81/universal-centre-of-sport
https://www.ucas.com/explore/unis/4ddd9e94/university-academy-92-ua92
https://www.ucas.com/explore/unis/9fed7298/university-centre-calderdale-college
https://www.ucas.com/explore/unis/33779918/university-centre-grimsby
https://www.ucas.com/explore/unis/554e082a/university-centre-rotherham
https://www.ucas.com/explore/unis/3c63c31c/university-centre-south-devon-south-devon-college
https://www.ucas.com/explore/unis/7e3d3c04/university-centre-south-essex
https://www.ucas.com/explore/unis/38a79022/university-centre-at-the-heart-of-yorkshire-education-group-inc-wakefield-college-selby-college
https://www.ucas.com/explore/unis/3c2fd5ff/university-college-of-estate-management
https://www.ucas.com/explore/unis/addf7cfa/uco-school-of-osteopathy-formerly-university-college-of-osteopathy
https://www.ucas.com/explore/unis/ba7a50f1/university-of-essex-online
https://www.ucas.com/explore/unis/0a4baa82/university-of-london-institute-in-paris
https://www.ucas.com/explore/unis/38236ddf/university-of-staffordshire
https://www.ucas.com/explore/unis/9928464e/university-of-the-arts-london
https://www.ucas.com/explore/unis/0ad2bf41/university-of-westminster-london
https://www.ucas.com/explore/unis/ff1a14b9/university-of-wales-trinity-saint-david
https://www.ucas.com/explore/unis/49afad53/walsall-college
https://www.ucas.com/explore/unis/a3e358dc/warburg-institute-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/e59d51a1/university-of-warwick
https://www.ucas.com/explore/unis/5d4edcb5/warwickshire-college-and-university-centre
https://www.ucas.com/explore/unis/3a9aee2e/weald-downland-living-museum
https://www.ucas.com/explore/unis/63b4cf1e/university-centre-west-anglia
https://www.ucas.com/explore/unis/54468f81/west-dean-college
https://www.ucas.com/explore/unis/a73de6e5/west-herts-college
https://www.ucas.com/explore/unis/4b00d284/university-of-west-london
https://www.ucas.com/explore/unis/d064e8a8/university-of-the-west-of-scotland
https://www.ucas.com/explore/unis/d157880f/west-suffolk-college
https://www.ucas.com/explore/unis/f8ee2f6a/capital-city-college-group
https://www.ucas.com/explore/unis/164a2e50/united-colleges-group
https://www.ucas.com/explore/unis/7c3e72d6/university-centre-weston
https://www.ucas.com/explore/unis/f845254e/weymouth-and-kingston-maurward-college
https://www.ucas.com/explore/unis/090716a0/wigan-and-leigh-college
https://www.ucas.com/explore/unis/37ab70cb/wiltshire-college-and-university-centre
https://www.ucas.com/explore/unis/34ba7b64/university-of-winchester
https://www.ucas.com/explore/unis/4940a04d/wirral-metropolitan-college
https://www.ucas.com/explore/unis/0a5af9c9/university-of-wolverhampton
https://www.ucas.com/explore/unis/fdb6a3ab/wrexham-university
https://www.ucas.com/explore/unis/4175f17f/xenophon-college-london
https://www.ucas.com/explore/unis/f4e4d70f/yeovil-college-university-centre
https://www.ucas.com/explore/unis/6045fd2c/york-college-university-centre
https://www.ucas.com/explore/unis/6d0cfb92/york-st-john-university
https://www.ucas.com/explore/unis/f40210c2/university-of-york"""

# Convert the university URLs from a string to a list
university_urls_list = university_urls.strip().split("\n")

# Function to ensure a string value or 'N/A' for empty fields
def safe_get_text(tag):
    if tag is None:
        return 'N/A'
    if isinstance(tag, str):
        return tag.strip()
    if isinstance(tag, (BeautifulSoup,)):
        return tag.get_text().strip()
    return str(tag).strip()  # Fall back to converting any non-string/non-BeautifulSoup type to string

# Function to scrape data from a university page
def scrape_university_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract the 'About Us' section
    about_section = soup.find('div', class_='font-prose')
    about_us = safe_get_text(about_section)

    # Extract the location
    location_section = soup.find('p', class_='header__subheading')
    location = safe_get_text(location_section)

    # Extract the official website link
    website_section = soup.find('a', {'data-link-type': 'website'})
    official_website = website_section['href'] if website_section and 'href' in website_section.attrs else 'N/A'

    # Extract the contact details (email and phone)
    contact_details_section = soup.find('footer', class_='content-block__section')
    email = 'N/A'
    phone = 'N/A'

    if contact_details_section:
        email_section = contact_details_section.find('a', href=lambda href: href and "mailto" in href)
        phone_section = contact_details_section.find('a', href=lambda href: href and "tel" in href)
        email = safe_get_text(email_section)
        phone = safe_get_text(phone_section)

    # Extract course locations
    course_locations_section = soup.find_all('article', class_='v5-card')
    course_locations = []
    for location in course_locations_section:
        location_name = safe_get_text(location.find('h2'))
        location_address = safe_get_text(location.find('div', class_='card__section'))
        course_locations.append({
            'location_name': location_name,
            'location_address': location_address
        })

    return {
        'url': url,
        'about_us': about_us,
        'location': location,
        'official_website': official_website,
        'contact_email': email,
        'contact_phone': phone,
        'course_locations': course_locations
    }

# List to hold data for all universities
universities_data = []

# Scrape data from each university URL
for url in university_urls_list:
    try:
        university_data = scrape_university_data(url)
        universities_data.append(university_data)
    except Exception as e:
        print(f"Error scraping {url}: {e}")

# Create a DataFrame from the data
df = pd.DataFrame(universities_data)

# Convert the DataFrame to JSON format
universities_json = df.to_json(orient='records', indent=4)

# Display the DataFrame and JSON output to the user

print("DataFrame successfully created and converted to JSON.")
import re

# Function to clean HTML tags from text
def clean_html(raw_html):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', raw_html)

# Clean the data inside the DataFrame
df['about_us'] = df['about_us'].apply(clean_html)
df['contact_email'] = df['contact_email'].apply(clean_html)
df['contact_phone'] = df['contact_phone'].apply(clean_html)
df['course_locations'] = df['course_locations'].apply(lambda x: [
    {'location_name': clean_html(loc['location_name']), 'location_address': clean_html(loc['location_address'])}
    for loc in x
])

# Convert the cleaned DataFrame to JSON format
cleaned_universities_json = df.to_json(orient='records', indent=4)

# Save the cleaned JSON data to a file
with open('cleaned_universities_data.json', 'w') as file:
    file.write(cleaned_universities_json)

print("Cleaned data saved to 'cleaned_universities_data.json'.")


```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# String containing all university URLs, separated by newlines
university_urls = """https://www.ucas.com/explore/unis/3d3a3bb5/acm-the-academy-of-contemporary-music
https://www.ucas.com/explore/unis/4e978513/arts-university-bournemouth
https://www.ucas.com/explore/unis/3824407c/associated-studios-performing-arts-academy
https://www.ucas.com/explore/unis/b7599006/university-college-birmingham
https://www.ucas.com/explore/unis/898276b4/bloomsbury-institute-london
https://www.ucas.com/explore/unis/627021f2/university-of-bradford
https://www.ucas.com/explore/unis/27ea792a/university-of-brighton
https://www.ucas.com/explore/unis/b79a4cbd/bristol-university-of-the-west-of-england
https://www.ucas.com/explore/unis/42121682/brunel-university-london
https://www.ucas.com/explore/unis/dfed9a6a/coventry-university
https://www.ucas.com/explore/unis/1de3729b/university-for-the-creative-arts
https://www.ucas.com/explore/unis/008ac8a8/university-of-derby
https://www.ucas.com/explore/unis/44f558c0/edge-hill-university
https://www.ucas.com/explore/unis/e4c66810/university-of-glasgow
https://www.ucas.com/explore/unis/e2cc6194/university-of-hertfordshire
https://www.ucas.com/explore/unis/c1c7b933/university-of-the-highlands-and-islands-uhi
https://www.ucas.com/explore/unis/31d5869a/hult-international-business-school
https://www.ucas.com/explore/unis/52087408/imperial-college-london
https://www.ucas.com/explore/unis/6cadf6e5/the-university-of-law
https://www.ucas.com/explore/unis/edaad69a/university-of-leicester
https://www.ucas.com/explore/unis/96faf64f/libf
https://www.ucas.com/explore/unis/970c6a07/university-of-lincoln
https://www.ucas.com/explore/unis/3e00a545/london-interdisciplinary-school-lis
https://www.ucas.com/explore/unis/bc261a05/mctimoney-college-of-chiropractic
https://www.ucas.com/explore/unis/5692045c/middlesex-university
https://www.ucas.com/explore/unis/9528a5b4/newcastle-university
https://www.ucas.com/explore/unis/1f6ba907/university-of-northampton
https://www.ucas.com/explore/unis/9698b0cc/nottingham-trent-university
https://www.ucas.com/explore/unis/79f42328/university-of-nottingham
https://www.ucas.com/explore/unis/a3bf2ac2/plymouth-marjon-university
https://www.ucas.com/explore/unis/250ab332/university-of-plymouth
https://www.ucas.com/explore/unis/ee9a8e04/pointblank-music-school
https://www.ucas.com/explore/unis/712655f4/richmond-american-university-london
https://www.ucas.com/explore/unis/895f238c/university-of-roehampton
https://www.ucas.com/explore/unis/f2b98edd/royal-veterinary-college-university-of-london
https://www.ucas.com/explore/unis/a33ecb69/university-of-south-wales
https://www.ucas.com/explore/unis/0f5d882f/university-of-southampton
https://www.ucas.com/explore/unis/f1f9a1ea/sp-jain-london-school-of-management
https://www.ucas.com/explore/unis/cf271b96/university-of-suffolk
https://www.ucas.com/explore/unis/6e856d76/university-of-sussex
https://www.ucas.com/explore/unis/4b70e003/swansea-university
https://www.ucas.com/explore/unis/62431d8f/university-of-worcester
https://www.ucas.com/explore/unis/3a023828/university-of-aberdeen
https://www.ucas.com/explore/unis/1558119a/abertay-university
https://www.ucas.com/explore/unis/168d2cd8/aberystwyth-university
https://www.ucas.com/explore/unis/a04f3f5e/academy-of-live-technology
https://www.ucas.com/explore/unis/1630c359/activate-learning
https://www.ucas.com/explore/unis/57f3e439/amity-university-in-london
https://www.ucas.com/explore/unis/d890f3a3/anglia-ruskin-university
https://www.ucas.com/explore/unis/15f49bd0/architectural-association-school-of-architecture
https://www.ucas.com/explore/unis/8554563c/arden-university
https://www.ucas.com/explore/unis/d4a8f435/arts-educational-schools
https://www.ucas.com/explore/unis/fd337b6e/arts-university-plymouth
https://www.ucas.com/explore/unis/95371765/aru-london
https://www.ucas.com/explore/unis/7f5fefd6/aru-writtle-formerly-writtle-university-college
https://www.ucas.com/explore/unis/7d51ec43/university-centre-askham-bryan
https://www.ucas.com/explore/unis/89c29ffe/aston-university-birmingham
https://www.ucas.com/explore/unis/7522cc92/bangor-university
https://www.ucas.com/explore/unis/a24b2b81/barnet-and-southgate-college
https://www.ucas.com/explore/unis/5f7d6526/barnfield-college-luton
https://www.ucas.com/explore/unis/4f52efe2/barnsley-college-university-centre
https://www.ucas.com/explore/unis/4c5e93f0/bath-college
https://www.ucas.com/explore/unis/29a67da9/bath-spa-university
https://www.ucas.com/explore/unis/295cc6a3/university-of-bath
https://www.ucas.com/explore/unis/a6fa6f7f/bedford-college-group
https://www.ucas.com/explore/unis/3261960f/university-of-bedfordshire
https://www.ucas.com/explore/unis/a537a0df/belfast-school-of-theology
https://www.ucas.com/explore/unis/5871db48/bimm-university
https://www.ucas.com/explore/unis/a5e9ffc8/birkbeck-university-of-london
https://www.ucas.com/explore/unis/24d90a06/birmingham-city-university
https://www.ucas.com/explore/unis/fa8f4026/bmet-birmingham-metropolitan-college
https://www.ucas.com/explore/unis/11ac4ef6/birmingham-newman-university
https://www.ucas.com/explore/unis/1d61dc76/university-of-birmingham
https://www.ucas.com/explore/unis/2519869b/university-centre-bishop-burton
https://www.ucas.com/explore/unis/a734fa1a/bishop-grosseteste-university
https://www.ucas.com/explore/unis/c2fe0703/blackburn-college
https://www.ucas.com/explore/unis/15d7c39f/blackpool-and-the-fylde-college
https://www.ucas.com/explore/unis/88e1a042/university-of-bolton
https://www.ucas.com/explore/unis/96f866ad/boomsatsuma
https://www.ucas.com/explore/unis/61cf12e4/boston-college
https://www.ucas.com/explore/unis/d4b5280f/bournemouth-and-poole-college
https://www.ucas.com/explore/unis/22e93c90/bournemouth-university
https://www.ucas.com/explore/unis/95d4242b/bpp-university
https://www.ucas.com/explore/unis/90e9d147/bradford-college
https://www.ucas.com/explore/unis/d849a737/bridgend-college
https://www.ucas.com/explore/unis/93e3e3b3/university-centre-somerset---bridgwater-taunton-college
https://www.ucas.com/explore/unis/0bdd99ec/brighton-and-sussex-medical-school
https://www.ucas.com/explore/unis/692982bf/city-of-bristol-college
https://www.ucas.com/explore/unis/da03c49a/trinity-college-bristol
https://www.ucas.com/explore/unis/69690f14/university-of-bristol
https://www.ucas.com/explore/unis/4cefe4e9/british-academy-of-jewellery
https://www.ucas.com/explore/unis/98e33cb7/bcom-eso---part-of-bcno-group-london-kent
https://www.ucas.com/explore/unis/1386af38/university-of-buckingham
https://www.ucas.com/explore/unis/592fda96/buckinghamshire-new-university
https://www.ucas.com/explore/unis/b0175104/burnley-college
https://www.ucas.com/explore/unis/74de5b4c/bury-college
https://www.ucas.com/explore/unis/27dbeccf/cambridge-marketing-college
https://www.ucas.com/explore/unis/3c88a10a/cambridge-regional-college
https://www.ucas.com/explore/unis/83b36cda/cambridge-school-of-visual-and-performing-arts
https://www.ucas.com/explore/unis/badc3603/university-of-cambridge
https://www.ucas.com/explore/unis/24bec0e2/canterbury-christ-church-university
https://www.ucas.com/explore/unis/6a3d68d5/cardiff-and-vale-college
https://www.ucas.com/explore/unis/777de2ea/cardiff-metropolitan-university
https://www.ucas.com/explore/unis/e1c1ad69/cardiff-university
https://www.ucas.com/explore/unis/016d384d/central-film-school-london
https://www.ucas.com/explore/unis/14062c40/university-of-central-lancashire-london
https://www.ucas.com/explore/unis/6533bebc/university-of-central-lancashire
https://www.ucas.com/explore/unis/81f409fb/centre-for-alternative-technology
https://www.ucas.com/explore/unis/09d21dac/university-of-chester
https://www.ucas.com/explore/unis/02675d55/chesterfield-college
https://www.ucas.com/explore/unis/52535f33/chichester-college-group-inc-brinsbury-chichester-crawley-northbrook-worthing-colleges
https://www.ucas.com/explore/unis/b1f140f3/university-of-chichester
https://www.ucas.com/explore/unis/bb82986c/city-guilds-of-london-art-school
https://www.ucas.com/explore/unis/e0a6cb6c/city-college-norwich-incorporating-easton-college
https://www.ucas.com/explore/unis/0e86a0cf/city-university-of-london
https://www.ucas.com/explore/unis/8ff764e4/cliff-college
https://www.ucas.com/explore/unis/c58b1306/university-centre-colchester-at-colchester-institute
https://www.ucas.com/explore/unis/12a66ed8/grwp-llandrillo-menai
https://www.ucas.com/explore/unis/1729a34a/collective-acting-studio
https://www.ucas.com/explore/unis/eaafa9fa/college-of-agriculture-food-and-rural-enterprise
https://www.ucas.com/explore/unis/878ef093/college-of-esports
https://www.ucas.com/explore/unis/a435d80e/cond-nast-college-of-fashion-design
https://www.ucas.com/explore/unis/6fc7541d/cornwall-college
https://www.ucas.com/explore/unis/6ec9c892/courtauld-institute-of-art-university-of-london
https://www.ucas.com/explore/unis/e86a79e3/coventry-college
https://www.ucas.com/explore/unis/2a9fba46/cranfield-university
https://www.ucas.com/explore/unis/aecda516/craven-college
https://www.ucas.com/explore/unis/ca2a0ff9/croydon-university-centre
https://www.ucas.com/explore/unis/ee20f1e0/university-of-cumbria
https://www.ucas.com/explore/unis/e859e326/david-game-college---higher-education-centre
https://www.ucas.com/explore/unis/11b2cb24/dbs-institute
https://www.ucas.com/explore/unis/6d06ed34/de-montfort-university
https://www.ucas.com/explore/unis/1fd8988e/derby-college
https://www.ucas.com/explore/unis/a3e4c2d4/dn-colleges-group
https://www.ucas.com/explore/unis/0299fcf2/dudley-college
https://www.ucas.com/explore/unis/bfcb0d31/university-of-dundee
https://www.ucas.com/explore/unis/6e455568/durham-university
https://www.ucas.com/explore/unis/2daba510/new-college-durham
https://www.ucas.com/explore/unis/70100535/dyslexia-action
https://www.ucas.com/explore/unis/7dce1b37/east-15-acting-school
https://www.ucas.com/explore/unis/59590204/university-of-east-anglia-uea
https://www.ucas.com/explore/unis/ae6f4d85/ekc-group
https://www.ucas.com/explore/unis/6d89f913/university-of-east-london
https://www.ucas.com/explore/unis/7837e9ef/east-riding-college
https://www.ucas.com/explore/unis/9597b298/east-sussex-college
https://www.ucas.com/explore/unis/dc3ccbfc/edinburgh-napier-university
https://www.ucas.com/explore/unis/114bb2e7/the-university-of-edinburgh
https://www.ucas.com/explore/unis/c615f773/escape-studios
https://www.ucas.com/explore/unis/24524819/escp-business-school
https://www.ucas.com/explore/unis/91f1f141/university-of-essex
https://www.ucas.com/explore/unis/ed93b6a5/european-school-of-economics
https://www.ucas.com/explore/unis/1f21f18b/exeter-college
https://www.ucas.com/explore/unis/91fd2fa0/university-of-exeter
https://www.ucas.com/explore/unis/13fe6f4f/falmouth-university
https://www.ucas.com/explore/unis/56ce8405/university-centre-farnborough
https://www.ucas.com/explore/unis/9a25a2cc/fashion-retail-academy
https://www.ucas.com/explore/unis/e1e76de4/furness-college
https://www.ucas.com/explore/unis/1018280f/futureworks
https://www.ucas.com/explore/unis/2304d818/gateshead-college
https://www.ucas.com/explore/unis/5cffcbbb/glasgow-caledonian-university
https://www.ucas.com/explore/unis/dd2bb28c/city-of-glasgow-college
https://www.ucas.com/explore/unis/a7656ba9/glasgow-school-of-art
https://www.ucas.com/explore/unis/098e22b9/gloucestershire-college
https://www.ucas.com/explore/unis/aa1a1601/university-of-gloucestershire
https://www.ucas.com/explore/unis/a8aae742/goldsmiths-university-of-london
https://www.ucas.com/explore/unis/d36d615d/gower-college-swansea
https://www.ucas.com/explore/unis/2c82b225/university-of-greenwich
https://www.ucas.com/explore/unis/5e389d01/guildhall-school-of-music-drama
https://www.ucas.com/explore/unis/87fbf767/halesowen-college
https://www.ucas.com/explore/unis/c9a6da19/harlow-college
https://www.ucas.com/explore/unis/a249ae49/harper-adams-university
https://www.ucas.com/explore/unis/788cb34b/harper-and-keele-veterinary-school
https://www.ucas.com/explore/unis/abf8e17f/hartpury-university
https://www.ucas.com/explore/unis/3ed8539f/havant-and-south-downs-college
https://www.ucas.com/explore/unis/9bca8f42/new-city-college-incorporating-havering-college
https://www.ucas.com/explore/unis/e9d03a65/haybridge-alliance-school-centred-initial-teacher-training
https://www.ucas.com/explore/unis/25f59ca8/harrow-richmond-uxbridge-college
https://www.ucas.com/explore/unis/d920c4e2/health-sciences-university-formerly-aecc-university-college
https://www.ucas.com/explore/unis/48a4bc2c/heart-of-worcestershire-college
https://www.ucas.com/explore/unis/ebb4f05e/hereford-college-of-arts
https://www.ucas.com/explore/unis/09715d18/herefordshire-ludlow-north-shropshire-college
https://www.ucas.com/explore/unis/5c658fd1/heriot-watt-university
https://www.ucas.com/explore/unis/73efa26e/hertford-regional-college
https://www.ucas.com/explore/unis/9ccfa755/holy-cross-sixth-form-college-and-university-centre
https://www.ucas.com/explore/unis/32bce31b/hopwood-hall-college
https://www.ucas.com/explore/unis/c666f13c/university-of-huddersfield
https://www.ucas.com/explore/unis/1231e6ce/hugh-baird-college
https://www.ucas.com/explore/unis/cbb9695b/hull-college
https://www.ucas.com/explore/unis/6a15243b/hull-york-medical-school
https://www.ucas.com/explore/unis/ac1fb4b5/university-of-hull
https://www.ucas.com/explore/unis/4215e8c6/icon-college-of-technology-and-management
https://www.ucas.com/explore/unis/9fd86a3e/institute-of-advanced-legal-studies-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/44eed984/institute-of-cancer-research-university-of-london
https://www.ucas.com/explore/unis/d04f617a/institute-of-commonwealth-studies-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/b829718a/institute-of-contemporary-music-performance
https://www.ucas.com/explore/unis/5737f385/institute-of-development-studies
https://www.ucas.com/explore/unis/909e0405/institute-of-english-studies-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/3e0010c4/institute-of-historical-research-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/08f1a8cc/institute-of-languages-cultures-and-societies-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/54791c19/international-business-college-manchester
https://www.ucas.com/explore/unis/16963c21/international-college-of-musical-theatre-icmt
https://www.ucas.com/explore/unis/a96bc33f/international-luxury-academy
https://www.ucas.com/explore/unis/c1eb45fd/irish-baptist-college
https://www.ucas.com/explore/unis/dd033310/islamic-college-for-advanced-studies
https://www.ucas.com/explore/unis/23aec7b0/istituto-marangoni-london
https://www.ucas.com/explore/unis/bb956d7d/jca-london-fashion-academy
https://www.ucas.com/explore/unis/5e07e833/keele-university
https://www.ucas.com/explore/unis/99fa626a/kent-and-medway-medical-school
https://www.ucas.com/explore/unis/aa22c216/university-of-kent
https://www.ucas.com/explore/unis/3cb12f40/king-stage
https://www.ucas.com/explore/unis/5fd4b1cc/king-s-college-london-university-of-london
https://www.ucas.com/explore/unis/58bbc794/kingston-maurward-college
https://www.ucas.com/explore/unis/4d6287f3/kingston-university
https://www.ucas.com/explore/unis/eb187029/klc-school-of-design
https://www.ucas.com/explore/unis/73bc7b26/lamda-london-academy-of-music-dramatic-art
https://www.ucas.com/explore/unis/058545dd/lancaster-university
https://www.ucas.com/explore/unis/b79f887e/lccm
https://www.ucas.com/explore/unis/272be5f0/leeds-arts-university
https://www.ucas.com/explore/unis/059b84a8/leeds-beckett-university
https://www.ucas.com/explore/unis/fe767449/university-centre-leeds-leeds-city-college
https://www.ucas.com/explore/unis/74edaea7/leeds-conservatoire
https://www.ucas.com/explore/unis/10bb36d7/leeds-trinity-university
https://www.ucas.com/explore/unis/b03b9d06/university-of-leeds
https://www.ucas.com/explore/unis/082a9246/leicester-college
https://www.ucas.com/explore/unis/d680310d/lincoln-college-university-centre
https://www.ucas.com/explore/unis/bf0964d9/city-of-liverpool-college-university-centre
https://www.ucas.com/explore/unis/78c33534/liverpool-hope-university
https://www.ucas.com/explore/unis/0097ca6a/liverpool-institute-for-performing-arts
https://www.ucas.com/explore/unis/b4d790d4/liverpool-john-moores-university
https://www.ucas.com/explore/unis/ff27f6b2/liverpool-school-of-tropical-medicine
https://www.ucas.com/explore/unis/9a2e99c3/university-of-liverpool
https://www.ucas.com/explore/unis/812e7c31/lma
https://www.ucas.com/explore/unis/32d682dc/london-business-school-university-of-london
https://www.ucas.com/explore/unis/f0896af7/london-college
https://www.ucas.com/explore/unis/0115c340/london-college-of-business-studies
https://www.ucas.com/explore/unis/d9ea977d/london-college-of-contemporary-arts
https://www.ucas.com/explore/unis/436d2160/london-film-academy
https://www.ucas.com/explore/unis/04dc593c/london-film-school
https://www.ucas.com/explore/unis/36dc6feb/london-metropolitan-university
https://www.ucas.com/explore/unis/25fe3db7/london-school-of-economics-and-political-science-university-of-london
https://www.ucas.com/explore/unis/b137fc85/london-school-of-hygiene-tropical-medicine-university-of-london
https://www.ucas.com/explore/unis/6ac2c56a/london-school-of-management-education
https://www.ucas.com/explore/unis/73c1c444/london-school-of-theology
https://www.ucas.com/explore/unis/b8a3b6cb/london-south-bank-university
https://www.ucas.com/explore/unis/aba77858/london-south-east-colleges-bromley-college
https://www.ucas.com/explore/unis/f838c19c/loughborough-college
https://www.ucas.com/explore/unis/b935bbb7/loughborough-university
https://www.ucas.com/explore/unis/bb9bf30f/manchester-metropolitan-university
https://www.ucas.com/explore/unis/01b03c44/university-of-manchester
https://www.ucas.com/explore/unis/e53c5903/medway-school-of-pharmacy
https://www.ucas.com/explore/unis/fa347416/metfilm
https://www.ucas.com/explore/unis/44c02f19/middlesbrough-college
https://www.ucas.com/explore/unis/7cb2e0fa/midkent-college-inc-medway-school-of-arts
https://www.ucas.com/explore/unis/a5ec7e73/morley-college-london
https://www.ucas.com/explore/unis/15cba371/moulton-college
https://www.ucas.com/explore/unis/3a59e89f/mountview-academy-of-theatre-arts
https://www.ucas.com/explore/unis/7d269cca/university-centre-myerscough
https://www.ucas.com/explore/unis/3886d908/national-college-of-hypnosis-psychotherapy
https://www.ucas.com/explore/unis/e0e252b0/national-film-and-television-school
https://www.ucas.com/explore/unis/a30c9a10/nazarene-theological-college
https://www.ucas.com/explore/unis/3a1c51f4/nelson-and-colne-college-group
https://www.ucas.com/explore/unis/cb222c5f/nescot
https://www.ucas.com/explore/unis/a524ae42/new-model-institute-for-technology-and-engineering-nmite
https://www.ucas.com/explore/unis/bbb76d07/new-school-of-psychotherapy-and-counselling
https://www.ucas.com/explore/unis/41110e8a/newbold-college
https://www.ucas.com/explore/unis/6783d4af/newcastle-college-university-centre
https://www.ucas.com/explore/unis/3cc3eeb1/newham-college-of-further-education
https://www.ucas.com/explore/unis/881106e7/norland
https://www.ucas.com/explore/unis/08b3f282/north-hertfordshire-college
https://www.ucas.com/explore/unis/aad4cda5/north-kent-college
https://www.ucas.com/explore/unis/ca9f4d85/north-warwickshire-and-south-leicestershire-college
https://www.ucas.com/explore/unis/77bfcfb8/northeastern-university-london
https://www.ucas.com/explore/unis/ad1ad199/northern-college-of-acupuncture
https://www.ucas.com/explore/unis/590e6636/the-northern-school-of-art
https://www.ucas.com/explore/unis/bc655e18/northern-school-of-contemporary-dance
https://www.ucas.com/explore/unis/408b9be4/northumbria-university-newcastle
https://www.ucas.com/explore/unis/bd68226f/norwich-university-of-the-arts
https://www.ucas.com/explore/unis/319ba5b1/nottingham-college
https://www.ucas.com/explore/unis/d1045d7f/nptc-group
https://www.ucas.com/explore/unis/4f7a0e47/oak-hill-theological-college
https://www.ucas.com/explore/unis/8c7662bf/oaklands-college
https://www.ucas.com/explore/unis/d2e1adfe/university-campus-oldham
https://www.ucas.com/explore/unis/7f559a95/open-university
https://www.ucas.com/explore/unis/d66b3801/orbital-south-colleges-university-centre
https://www.ucas.com/explore/unis/d422b186/oxford-brookes-university
https://www.ucas.com/explore/unis/351374e0/university-of-oxford
https://www.ucas.com/explore/unis/5986cf35/peter-symonds-college
https://www.ucas.com/explore/unis/7ca2e52f/university-centre-peterborough
https://www.ucas.com/explore/unis/fba7f8e7/petroc
https://www.ucas.com/explore/unis/be7280d5/plumpton-college
https://www.ucas.com/explore/unis/babaeca8/city-college-plymouth
https://www.ucas.com/explore/unis/039b8331/university-of-portsmouth
https://www.ucas.com/explore/unis/403c3afa/queen-margaret-university-edinburgh
https://www.ucas.com/explore/unis/856d62c2/queen-mary-university-of-london
https://www.ucas.com/explore/unis/3e45ece0/queen-s-university-belfast
https://www.ucas.com/explore/unis/6f003a01/ravensbourne-university-london
https://www.ucas.com/explore/unis/c283b1b0/university-of-reading
https://www.ucas.com/explore/unis/b3301895/real-training
https://www.ucas.com/explore/unis/ae8f02f9/university-centre-reaseheath
https://www.ucas.com/explore/unis/4687847f/refugee-law-initiative-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/d22aa302/regent-s-university-london
https://www.ucas.com/explore/unis/60144100/regents-theological-college
https://www.ucas.com/explore/unis/5a3cac22/robert-gordon-university
https://www.ucas.com/explore/unis/5bc57c84/rose-bruford-college
https://www.ucas.com/explore/unis/94090b0b/royal-academy-of-dance
https://www.ucas.com/explore/unis/ac1ac95e/royal-academy-of-dramatic-art
https://www.ucas.com/explore/unis/6650b619/royal-academy-of-music-university-of-london
https://www.ucas.com/explore/unis/0014c21f/royal-agricultural-university
https://www.ucas.com/explore/unis/e4015e94/royal-birmingham-conservatoire
https://www.ucas.com/explore/unis/098ac3d1/royal-central-school-of-speech-and-drama-university-of-london
https://www.ucas.com/explore/unis/b6e2e7ed/royal-college-of-art
https://www.ucas.com/explore/unis/5f0445dd/royal-college-of-music
https://www.ucas.com/explore/unis/22072435/royal-conservatoire-of-scotland
https://www.ucas.com/explore/unis/fb1d3c04/royal-drawing-school
https://www.ucas.com/explore/unis/5b13b3ae/royal-holloway-university-of-london
https://www.ucas.com/explore/unis/dd590a4b/royal-northern-college-of-music
https://www.ucas.com/explore/unis/76a6ba43/royal-welsh-college-of-music-and-drama
https://www.ucas.com/explore/unis/db037e28/sae-institute
https://www.ucas.com/explore/unis/81fe3292/university-of-st-andrews
https://www.ucas.com/explore/unis/9b3207d7/st-george-s-university-of-london
https://www.ucas.com/explore/unis/e47a9792/st-mary-s-university-college-belfast
https://www.ucas.com/explore/unis/8432fe0e/university-centre-at-salford-city-college
https://www.ucas.com/explore/unis/07fe4a70/university-of-salford
https://www.ucas.com/explore/unis/80031aa0/sgs-south-gloucestershire-and-stroud-college
https://www.ucas.com/explore/unis/86bcfbc8/sheffield-college
https://www.ucas.com/explore/unis/af611462/sheffield-hallam-university
https://www.ucas.com/explore/unis/dbd76d8b/university-of-sheffield
https://www.ucas.com/explore/unis/a840508b/shrewsbury-colleges-group
https://www.ucas.com/explore/unis/2a479d97/coleg-sir-gar
https://www.ucas.com/explore/unis/01d1372f/smb-college-group
https://www.ucas.com/explore/unis/63840f6d/soas-university-of-london
https://www.ucas.com/explore/unis/26360c43/solihull-college-university-centre
https://www.ucas.com/explore/unis/b532a77f/sotheby-s-institute-of-art-london
https://www.ucas.com/explore/unis/0c59f8f5/south-and-city-college-birmingham
https://www.ucas.com/explore/unis/b44ad48d/south-thames-college-group-inc-carshalton-south-thames-and-kingston-college
https://www.ucas.com/explore/unis/f92b2692/solent-university-southampton
https://www.ucas.com/explore/unis/e71cb3c4/university-centre-sparsholt
https://www.ucas.com/explore/unis/dc96ef3c/spurgeon-s-college
https://www.ucas.com/explore/unis/ba944ca6/sruc-scotland-s-rural-college
https://www.ucas.com/explore/unis/f827e91c/university-centre-st-helens
https://www.ucas.com/explore/unis/1264b52a/st-mary-s-university-twickenham
https://www.ucas.com/explore/unis/20b2137b/university-of-stirling
https://www.ucas.com/explore/unis/337ec3b8/stranmillis-university-college
https://www.ucas.com/explore/unis/5cc768aa/university-of-strathclyde
https://www.ucas.com/explore/unis/31260746/city-of-sunderland-college
https://www.ucas.com/explore/unis/5f968715/university-of-sunderland
https://www.ucas.com/explore/unis/5ea13bf3/university-of-surrey
https://www.ucas.com/explore/unis/b35862e3/the-college-swansea-university
https://www.ucas.com/explore/unis/16a7e353/new-college-swindon-university-centre
https://www.ucas.com/explore/unis/4156d693/tameside-college
https://www.ucas.com/explore/unis/bffe1f14/the-engineering-design-institute-london
https://www.ucas.com/explore/unis/b46d1424/teesside-university
https://www.ucas.com/explore/unis/6c26b7ba/thinkspace-education
https://www.ucas.com/explore/unis/ad154893/trafford-stockport-college-group
https://www.ucas.com/explore/unis/09c81910/trinity-laban
https://www.ucas.com/explore/unis/cfe1b520/university-centre-truro-and-penwith
https://www.ucas.com/explore/unis/117bff52/two-mile-ash-ittp-partnered-with-chiltern-training-group
https://www.ucas.com/explore/unis/8df995e9/tyne-coast-college
https://www.ucas.com/explore/unis/babb7b8c/ucen-manchester
https://www.ucas.com/explore/unis/9bebda25/ucfb
https://www.ucas.com/explore/unis/bd0c70d8/ucl-university-college-london
https://www.ucas.com/explore/unis/b0c96a45/ulster-university
https://www.ucas.com/explore/unis/2f918b81/universal-centre-of-sport
https://www.ucas.com/explore/unis/4ddd9e94/university-academy-92-ua92
https://www.ucas.com/explore/unis/9fed7298/university-centre-calderdale-college
https://www.ucas.com/explore/unis/33779918/university-centre-grimsby
https://www.ucas.com/explore/unis/554e082a/university-centre-rotherham
https://www.ucas.com/explore/unis/3c63c31c/university-centre-south-devon-south-devon-college
https://www.ucas.com/explore/unis/7e3d3c04/university-centre-south-essex
https://www.ucas.com/explore/unis/38a79022/university-centre-at-the-heart-of-yorkshire-education-group-inc-wakefield-college-selby-college
https://www.ucas.com/explore/unis/3c2fd5ff/university-college-of-estate-management
https://www.ucas.com/explore/unis/addf7cfa/uco-school-of-osteopathy-formerly-university-college-of-osteopathy
https://www.ucas.com/explore/unis/ba7a50f1/university-of-essex-online
https://www.ucas.com/explore/unis/0a4baa82/university-of-london-institute-in-paris
https://www.ucas.com/explore/unis/38236ddf/university-of-staffordshire
https://www.ucas.com/explore/unis/9928464e/university-of-the-arts-london
https://www.ucas.com/explore/unis/0ad2bf41/university-of-westminster-london
https://www.ucas.com/explore/unis/ff1a14b9/university-of-wales-trinity-saint-david
https://www.ucas.com/explore/unis/49afad53/walsall-college
https://www.ucas.com/explore/unis/a3e358dc/warburg-institute-school-of-advanced-study-university-of-london
https://www.ucas.com/explore/unis/e59d51a1/university-of-warwick
https://www.ucas.com/explore/unis/5d4edcb5/warwickshire-college-and-university-centre
https://www.ucas.com/explore/unis/3a9aee2e/weald-downland-living-museum
https://www.ucas.com/explore/unis/63b4cf1e/university-centre-west-anglia
https://www.ucas.com/explore/unis/54468f81/west-dean-college
https://www.ucas.com/explore/unis/a73de6e5/west-herts-college
https://www.ucas.com/explore/unis/4b00d284/university-of-west-london
https://www.ucas.com/explore/unis/d064e8a8/university-of-the-west-of-scotland
https://www.ucas.com/explore/unis/d157880f/west-suffolk-college
https://www.ucas.com/explore/unis/f8ee2f6a/capital-city-college-group
https://www.ucas.com/explore/unis/164a2e50/united-colleges-group
https://www.ucas.com/explore/unis/7c3e72d6/university-centre-weston
https://www.ucas.com/explore/unis/f845254e/weymouth-and-kingston-maurward-college
https://www.ucas.com/explore/unis/090716a0/wigan-and-leigh-college
https://www.ucas.com/explore/unis/37ab70cb/wiltshire-college-and-university-centre
https://www.ucas.com/explore/unis/34ba7b64/university-of-winchester
https://www.ucas.com/explore/unis/4940a04d/wirral-metropolitan-college
https://www.ucas.com/explore/unis/0a5af9c9/university-of-wolverhampton
https://www.ucas.com/explore/unis/fdb6a3ab/wrexham-university
https://www.ucas.com/explore/unis/4175f17f/xenophon-college-london
https://www.ucas.com/explore/unis/f4e4d70f/yeovil-college-university-centre
https://www.ucas.com/explore/unis/6045fd2c/york-college-university-centre
https://www.ucas.com/explore/unis/6d0cfb92/york-st-john-university
https://www.ucas.com/explore/unis/f40210c2/university-of-york"""

# Convert the university URLs from a string to a list
university_urls_list = university_urls.strip().split("\n")

# Function to ensure a string value or 'N/A' for empty fields
def safe_get_text(tag):
    if tag is None:
        return 'N/A'
    if isinstance(tag, str):
        return tag.strip()
    if isinstance(tag, (BeautifulSoup,)):
        return tag.get_text().strip()
    return str(tag).strip()  # Fall back to converting any non-string/non-BeautifulSoup type to string

# Function to clean HTML
def clean_html(raw_html):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', raw_html)

# Function to scrape data from a university page
def scrape_university_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract the 'About Us' section
    about_section = soup.find('div', class_='font-prose')
    about_us = safe_get_text(about_section)

    # Extract the location
    location_section = soup.find('p', class_='header__subheading')
    if location_section is None:
        # Check if location is available in another tag format
        location_section = soup.find('p', attrs={'data-v-d356407d': True})
    location = safe_get_text(location_section)

    # Extract the official website link
    website_section = soup.find('a', {'data-link-type': 'website'})
    official_website = website_section['href'] if website_section and 'href' in website_section.attrs else 'N/A'

    # Extract the contact details (email and phone)
    contact_details_section = soup.find('footer', class_='content-block__section')
    email = 'N/A'
    phone = 'N/A'

    if contact_details_section:
        email_section = contact_details_section.find('a', href=lambda href: href and "mailto" in href)
        phone_section = contact_details_section.find('a', href=lambda href: href and "tel" in href)
        email = safe_get_text(email_section)
        phone = safe_get_text(phone_section)

    # Extract course locations
    course_locations_section = soup.find_all('article', class_='v5-card')
    course_locations = []
    for location in course_locations_section:
        location_name = safe_get_text(location.find('h2'))
        location_address = safe_get_text(location.find('div', class_='card__section'))
        course_locations.append({
            'location_name': location_name,
            'location_address': location_address
        })

    # Extract the logo image URL
    logo_section = soup.find('div', class_='header__logo')
    logo_url = logo_section.find('img')['src'] if logo_section and logo_section.find('img') else 'N/A'

    # Extract the background image URL
    header_section = soup.find('header', class_='v5-header')
    background_image = header_section['style'].split('url(')[-1].split(')')[0] if header_section else 'N/A'

    # Extract the gallery images
    gallery_images_section = soup.find_all('div', class_='swiper-slide')
    gallery_images = [img['style'].split('url("')[1].split('")')[0] for img in gallery_images_section if 'background-image' in img['style']]

    return {
        'url': url,
        'about_us': about_us,
        'location': location,
        'official_website': official_website,
        'contact_email': email,
        'contact_phone': phone,
        'course_locations': course_locations,
        'logo_url': logo_url,
        'background_image': background_image,
        'gallery_images': gallery_images
    }

# List to hold data for all universities
universities_data = []

# Scrape data from each university URL
for url in university_urls_list:
    try:
        university_data = scrape_university_data(url)
        universities_data.append(university_data)
    except Exception as e:
        print(f"Error scraping {url}: {e}")

# Create a DataFrame from the data
df = pd.DataFrame(universities_data)

# Clean the data inside the DataFrame
df['about_us'] = df['about_us'].apply(clean_html)
df['contact_email'] = df['contact_email'].apply(clean_html)
df['contact_phone'] = df['contact_phone'].apply(clean_html)
df['course_locations'] = df['course_locations'].apply(lambda x: [
    {'location_name': clean_html(loc['location_name']), 'location_address': clean_html(loc['location_address'])}
    for loc in x
])

# Convert the DataFrame to JSON format
cleaned_universities_json = df.to_json(orient='records', indent=4)

# Save the cleaned JSON data to a file
with open('cleaned_universities_with_images_data.json', 'w') as file:
    file.write(cleaned_universities_json)

print("Cleaned data with images saved to 'cleaned_universities_with_images_data.json'.")
```

```python
import json
import re

# Function to clean HTML tags from text
def clean_html(raw_html):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', raw_html)

# Function to read data from a JSON file
def read_json_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data

# Function to clean the university data by removing HTML tags
def clean_university_data(data):
    for entry in data:
        entry['university_name'] = clean_html(entry['university_name'])
        entry['location'] = clean_html(entry['location'])
        entry['official_website'] = clean_html(entry['official_website'])
        entry['contact_email'] = clean_html(entry['contact_email'])
        entry['contact_phone'] = clean_html(entry['contact_phone'])

        # Clean course locations
        for loc in entry['course_locations']:
            loc['location_name'] = clean_html(loc['location_name'])
            loc['location_address'] = clean_html(loc['location_address'])

        entry['logo_url'] = clean_html(entry['logo_url'])
        entry['background_image'] = clean_html(entry['background_image'])
        entry['about_us'] = clean_html(entry['about_us'])

    return data

# Function to write the cleaned data to a new JSON file
def write_cleaned_data_to_file(cleaned_data, output_file_path):
    with open(output_file_path, 'w', encoding='utf-8') as file:
        json.dump(cleaned_data, file, ensure_ascii=False, indent=4)

# Main function to process the file
def process_file(input_file_path, output_file_path):
    raw_data = read_json_from_file(input_file_path)
    cleaned_data = clean_university_data(raw_data)
    write_cleaned_data_to_file(cleaned_data, output_file_path)

# Specify input and output file paths
input_file_path = '/content/uni.json'  # Replace with your input file path
output_file_path = '/content/universities_details.json' # Specify the output file path

# Process the input file
process_file(input_file_path, output_file_path)

print(f"Cleaned data has been saved to '{output_file_path}'.")
```
