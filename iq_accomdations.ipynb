```python
import requests
from bs4 import BeautifulSoup
import json

# Base URL
base_url = "https://www.iqstudentaccommodation.com/"

# List of cities
cities = [
    "Bangor", "Bath", "Birmingham", "Bradford", "Brighton", "Bristol",
    "Coventry", "Dundee", "Edinburgh", "Egham", "Glasgow", "Huddersfield",
    "Kingston", "Leeds", "Leicester", "Lincoln", "Liverpool", "London",
    "Manchester", "Newcastle", "Nottingham", "Oxford", "Plymouth", "Preston",
    "Salford", "Sheffield", "Wolverhampton", "York"
]

# Headers to mimic a browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Function to scrape data from a single city page
def scrape_city(city):
    city_url = f"{base_url}{city.lower()}/"
    response = requests.get(city_url, headers=headers)

    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to retrieve data for {city}. Status code: {response.status_code}")
        return []

    # Debugging: Print the response text to verify we're getting the correct HTML
    print(f"HTML content for {city}:\n", response.text[:1000])  # print the first 1000 characters of the HTML

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all room cards on the page
    room_cards = soup.find_all('div', class_='sP5SLwX_FK8_SgOzKnOm')  # Adjust class as needed

    if not room_cards:
        print(f"No room cards found for {city}")

    accommodations = []

    for card in room_cards:
        try:
            # Extract image and room link
            image_tag = card.find('img', class_='Zw93tOFu9FsMPUSr6k63')
            image_url = image_tag['src'] if image_tag else "No Image"

            room_link_tag = card.find('a', {'data-testid': 'room-card-carousel-url-0'})
            room_link = base_url.rstrip('/') + room_link_tag['href'] if room_link_tag else "No Link"

            # Extract location
            location = card.find('div', {'data-testid': 'room-card-location'}).text.strip() if card.find('div', {'data-testid': 'room-card-location'}) else "No Location"

            # Extract room details
            room_name = card.find('a', {'data-testid': 'room-card-link'}).text.strip() if card.find('a', {'data-testid': 'room-card-link'}) else "No Room Name"

            # Extract additional room details (like room type and size)
            room_details = [detail.text for detail in card.find_all('div', class_='e8z3LP_fM4QSsknuebAI')]

            # Extract cost
            cost_tag = card.find('h6', {'data-testid': 'heading-6'})
            cost = cost_tag.text.strip() if cost_tag else "No Cost Info"

            accommodations.append({
                'city': city,
                'location': location,
                'room_name': room_name,
                'image_url': image_url,
                'room_link': room_link,
                'details': room_details,
                'cost': cost
            })
        except Exception as e:
            print(f"Error extracting data from {city}: {e}")

    return accommodations

# Store data in a JSON file
def save_to_json(data):
    with open('/content/acc100.json', 'w', encoding='utf-8') as file:
        json.dump(data, file, indent=4, ensure_ascii=False)

# Main scraping process
all_accommodations = []
for city in cities:
    city_accommodations = scrape_city(city)
    all_accommodations.extend(city_accommodations)
    print(f"Data extracted for {city}")

# Save data to JSON
save_to_json(all_accommodations)

print("Scraping complete. Data saved to accommodations.json.")
```

```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
import time
import json

# Function to initialize Firefox WebDriver
def init_driver():
    firefox_options = Options()
    firefox_options.add_argument("--headless")  # Run in headless mode
    firefox_options.add_argument("--no-sandbox")
    firefox_options.add_argument("--disable-dev-shm-usage")

    # Use Geckodriver (for Firefox)
    service = Service("/usr/bin/geckodriver")
    driver = webdriver.Firefox(service=service, options=firefox_options)
    return driver

# Base URL
base_url = "https://www.iqstudentaccommodation.com/"

# List of cities
cities = [
    "Bangor", "Bath", "Birmingham", "Bradford", "Brighton", "Bristol",
    "Coventry", "Dundee", "Edinburgh", "Egham", "Glasgow", "Huddersfield",
    "Kingston", "Leeds", "Leicester", "Lincoln", "Liverpool", "London",
    "Manchester", "Newcastle", "Nottingham", "Oxford", "Plymouth", "Preston",
    "Salford", "Sheffield", "Wolverhampton", "York"
]

# Function to scrape data from a single city page using Selenium
def scrape_city(city, driver):
    city_url = f"{base_url}{city.lower()}/"
    driver.get(city_url)
    time.sleep(5)  # Wait for the JavaScript to load

    accommodations = []

    try:
        # Find all room cards on the page (adjust the class name if needed)
        room_cards = driver.find_elements(By.CLASS_NAME, 'sP5SLwX_FK8_SgOzKnOm')  # Adjust this class as needed

        if not room_cards:
            print(f"No room cards found for {city}")
            return []

        for card in room_cards:
            try:
                # Extract image and room link
                image_tag = card.find_element(By.TAG_NAME, 'img')
                image_url = image_tag.get_attribute('src') if image_tag else "No Image"

                room_link_tag = card.find_element(By.TAG_NAME, 'a')
                room_link = base_url.rstrip('/') + room_link_tag.get_attribute('href') if room_link_tag else "No Link"

                # Extract location
                location = card.find_element(By.XPATH, "//div[@data-testid='room-card-location']").text.strip()

                # Extract room details
                room_name = card.find_element(By.XPATH, "//a[@data-testid='room-card-link']").text.strip()

                # Extract additional room details (like room type and size)
                room_details = [detail.text for detail in card.find_elements(By.CLASS_NAME, 'e8z3LP_fM4QSsknuebAI')]

                # Extract cost
                cost_tag = card.find_element(By.XPATH, "//h6[@data-testid='heading-6']")
                cost = cost_tag.text.strip() if cost_tag else "No Cost Info"

                accommodations.append({
                    'city': city,
                    'location': location,
                    'room_name': room_name,
                    'image_url': image_url,
                    'room_link': room_link,
                    'details': room_details,
                    'cost': cost
                })
            except Exception as e:
                print(f"Error extracting data from {city}: {e}")

    except Exception as e:
        print(f"Error scraping city {city}: {e}")

    return accommodations

# Store data in a JSON file
def save_to_json(data):
    with open('accommodations.json', 'w', encoding='utf-8') as file:
        json.dump(data, file, indent=4, ensure_ascii=False)

# Main scraping process
def main():
    all_accommodations = []
    driver = init_driver()

    try:
        for city in cities:
            city_accommodations = scrape_city(city, driver)
            all_accommodations.extend(city_accommodations)
            print(f"Data extracted for {city}")

    finally:
        driver.quit()  # Always close the driver when done

    # Save data to JSON
    save_to_json(all_accommodations)
    print("Scraping complete. Data saved to accommodations.json.")

# Run the script
if __name__ == "__main__":
    main()
```
